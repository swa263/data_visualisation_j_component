```python
## Load Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load dataset
# Discipline = Sport, NOC - National Olympic Committee
# Their role is to ensure the representation of their respective countries at the Olympic Games by sending competitors and 
# officials as well as to promote the fundamental principles and values of Olympism in their countries, 
# in particular in the fields of sport and education.
#
dataset = pd.read_excel("Dataset\Teams.xlsx", engine='openpyxl')
dataset.head()
```

```python
## Basic Information Of Dataset
#
dataset.info()
```

```python
## Statistics Information Of Dataset
#
dataset.describe()
```

```python
## Count unique values of all features in dataset
#
columns = dataset.describe().columns
unique_count = dataset.describe()[dataset.describe().index == 'unique'].values.flatten().tolist()

plt.bar(x = columns, height = unique_count)
plt.title("Visual Representation Of Unique Value In Each Features")
plt.xlabel("Features")
plt.ylabel("Unique value of each feature")
plt.show()
plt.close()
```

```python
## Teams Participated Count With Unique Sport
#
sns.countplot(data = dataset, x = 'Discipline')
plt.title("Count of unique sports/discipline")
plt.xlabel('Sports')
plt.ylabel('Counts')
plt.xticks(rotation='vertical')
plt.show()
plt.close()
```

```python
## Teams Participated Counts With Country
#
sns.countplot(data = dataset, x = 'Name')
plt.rc('figure', figsize=(30, 30))
plt.title("Count of Teams With Nation")
plt.xlabel('Names')
plt.ylabel('Counts')
plt.xticks(rotation='vertical')
plt.show()
plt.close()
```
